3. Merge these two DataFrames into a single DataFrame called census.



Please use the following dataset for merging purposes:


'https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/state-populations.csv'

'https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/census-divisions.csv'

```python
# codes here
population = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/state-populations.csv')
division = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/census-divisions.csv')

census = pd.merge(population, division, on="state")
census.dropna()
```

4. Re-shape census so that one column contains all population measures, and another the year attributes.



```python
# codes here

census_reshape = census
census_reshape.set_index(['state', 'region', 'division'], inplace=True)

census_reshape = census_reshape.stack().to_frame()
census_reshape.rename({0: 'population'}, axis=1, inplace=True)

census_reshape.index.names = [*census_reshape.index.names[0:3], 'year']

census_reshape
```

5. Group the data by year and summarize it.



```python
# codes here

census_groupby_year = census_reshape.groupby('year')
census_groupby_year
print("Population grouped by year:", census_groupby_year.sum().add_prefix('sum_'))
```

6. Group the data by region, division and year and summarize it.

```python
# codes here

census_groupby_region_division_year = census_reshape.groupby(['region','division','year'])
census_groupby_region_division_year
print("Population grouped by region, division, and year:", census_groupby_region_division_year.sum().add_prefix('sum_'))
```
