```python
#codes here for a)

import math
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt


def demo1():
    mu, sigma = 0, 0.1
    sampleNo = 1000
    s = np.random.normal(mu, sigma, sampleNo)
    plt.hist(s, bins=100, density=True)
    plt.show()

demo1()
```
